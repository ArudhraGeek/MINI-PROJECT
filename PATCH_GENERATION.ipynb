```
def create_image(rgb_img,h1,w1,h2,w2,ans):
    rootpath = create_image
    crop_img111 = "crop_img" + "001_" + ans
    crop_img1 = crop_img111 + ".png"
    crop_img11 = rgb_img[h1:h2, w1:w2]
    plt.figure()
    plt.imshow(crop_img11)
    #print(crop_img11.shape)
    cv2.imwrite(crop_img1, crop_img11)
```

```
img = cv2.imread("gdrive/My Drive/DATASET/image001.png")
```

```
#cv2.imshow('image',img)
b,g,r = cv2.split(img)       # get b,g,r
rgb_img = cv2.merge([r,g,b])
r_img = cv2.merge([r])
#g_img = cv2.merge([g])
# switch it to rgb
```

```
plt.figure()
plt.imshow(rgb_img)
plt.figure()
plt.imshow(r_img)
```

```
h,w = r_img.shape
#print(h)
#print(w)
```

```
#generate 400x400 patch from the image
import random
for i in range(0,20):
    h1 = random.randrange(0, h-400)
    w1 = random.randrange(0, w-400)
    #print(h1)
    #print(w1)
    h2 = h1+400
    w2 = w1+400
    create_image(r_img,h1,w1,h2,w2,str(i))
```
